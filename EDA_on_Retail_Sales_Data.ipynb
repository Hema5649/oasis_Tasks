```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Data Loading and Cleaning
retail_data = pd.read_csv('retail_sales_dataset.csv')
retail_data['Date'] = pd.to_datetime(retail_data['Date'])
retail_data.dropna(inplace=True)  # Assuming we drop rows with missing values
```

```python
# Descriptive Statistics
basic_stats = retail_data.describe()
# Time Series Analysis
plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='Total Amount', data=retail_data)
plt.title('Trend of Sales Over Time')
plt.xlabel('Order Date')
plt.ylabel('Sales')
plt.show()
```

```python
# Customer and Product Analysis
customer_demographics = retail_data.groupby('Gender').agg({'Age': ['mean', 'median'], 'Customer ID': 'count'})
product_analysis = retail_data.groupby('Product Category').agg({'Quantity': 'sum', 'Total Amount': 'sum'})
# Visualization
plt.figure(figsize=(10, 6))
sns.barplot(x=customer_demographics.index, y=customer_demographics[('Customer ID', 'count')])
plt.title('Customer Count by Gender')
plt.xlabel('Gender')
plt.ylabel('Customer Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=product_analysis.index, y=product_analysis['Total Amount'])
plt.title('Total Sales by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```
